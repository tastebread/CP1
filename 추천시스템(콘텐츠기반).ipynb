```python
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False
import numpy as np
import random
import mplcyberpunk
import pandas as pd
import seaborn as sns
import matplotlib as mpl

import warnings

%matplotlib inline
plt.style.use('cyberpunk')
warnings.filterwarnings("ignore")
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')

movies.shape
movies.head()
```

```python
#주요 컬럼만 추출해서 데이터 프레임을 새로 생성

col = ['id','title','genres','vote_average','vote_count','popularity','keywords','overview']
"""
id : 아이디
title : 영화 제목
genres : 영화 장르
vote_average,count : 평균평점, 평점 투표 수
popularity : 영화의 인기 정도
keywords : 영화를 설명하는 주요 키워드 문구
overview : 영화에 대한 개요 설명
"""
new_movies = movies[col]
new_movies
```

```python
pd.set_option('max_colwidth',80)
new_movies.head(1)
```

```python
from ast import literal_eval

#genres 컬럼은 리스트내의 사전인 것으로 실제로는 전체가 하나의 문자열로 되어있다
#literal_eval -> 딕셔너리 형태로 변경해준다 이렇게 되면 key-value 를 사용할 수 있음

#문자열을 객체로 변경 : 리스트 내의 사전 으로 변경
new_movies['genres'] = new_movies['genres'].apply(literal_eval) 
new_movies['keywords'] = new_movies['keywords'].apply(literal_eval)


#객체에서 name만 추출해서 올린다

new_movies['genres'] = new_movies['genres'].apply(lambda x : [ dic['name'] for dic in x] )
new_movies['keywords'] = new_movies['keywords'].apply(lambda x : [ dic['name'] for dic in x] )

new_movies[['genres','keywords']]
```

```python
#장르 콘텐츠 유사도 측정

from sklearn.feature_extraction.text import CountVectorizer

"""
CountVectorizer : 문서를 token count matrix 로 변환하는 클래스 이다
"""

#리스트의 객체를 다시 문자열로 변환
new_movies['literal_genres'] = new_movies['genres'].apply(lambda x: (' ').join(x))

#CountVectorizer
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))

genre_mat = count_vect.fit_transform(new_movies['literal_genres'])

print(genre_mat.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

"""
cosine_similarity 를 이용해 장르 유사도 행렬 생성
"""
genre_sim = cosine_similarity(genre_mat, genre_mat)
genre_sim[0]
```

```python
#특정 영화 와 장르 유사도가 높은 영화의 정보를 얻기 위한 함수를 생성

def find_sim_movie(df, sim_matrix, title_name, top_n=10):
    
    
    #입력한 영화의 index
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    #입력한 영화의 유사도 데이터 프레임 추카
    df['similarity'] = sim_matrix[title_index, :].reshape(-1,1)
    
    #유사도 내림차순 정렬 후 상위 index 추출
    temp = df.sort_values(by='similarity', ascending=False)
    final_index = temp.index.values[ : top_n]
    
    return df.iloc[final_index]
```

```python
#대부와 장르별 유사도가 높은 영화 10개

similar_movies = find_sim_movie(new_movies, genre_sim, 'The Godfather', 10)
similar_movies[['title','vote_average','similarity']]
```

```python
#평점이 높은 순으로 영화를 정렬함  10개의 Index 만 출력

new_movies[['title','vote_average','vote_count']].sort_values('vote_average',ascending=False)[:10]
```

```python
#가중 평점 함수 생성

percentile = 0.6

m = new_movies['vote_count'].quantile(percentile)
C = new_movies['vote_average'].mean()

def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ( (v/(v+m)) * R ) + ( (m/(m+v)) * C ) 


new_movies['weighted_vote'] = new_movies.apply(weighted_vote_average, axis=1)
```

```python
temp = new_movies[['title','vote_average','vote_count','weighted_vote']]
temp.sort_values('weighted_vote', ascending=False)[:10]
```
